## Example 1: Element uniqueness problem 

We want to know whether all n elements of a sequence are unique. 

1. Base case: if n=1, the elements are trivally unique
2. For n >=2, the elements are unique if and only if:
    1) the first n-1 are unqiue 
    2) the last n-1 are unique 
    3) the first and last elements are different 

```python
## A bad example 

def unique(S,start,stop):
    '''
    Return True if there are no duplicate elements in slices S[start:stop]

    '''

    # base case:
    if stop - start <=1:
        return True 
    elif not unique(S,start,stop-1):
        return False
    elif not unique(S,start+1,stop):
        return False
    else:
        return S[start]!= S[stop-1]
```

Each time of the function call would generate two call on the same function. Thehrefore, for the worst case, it would run 

1 + 2 + 4 + ...+ 2^(n-1) where n=stop-start.

Therefore, the running time of function is O(2^n) 

```python
## Exercise C-4.11
## An efficient recursive function for solving the element uniqueness problem, which runs in time that is at most O(n^2) in the worst case without using sorting. 

def unique_good(S,start,stop):
    if stop-start <=1:
        return True
    for index in range(stop-start):
        i = S[start+index]
        print(index,i)
        for j in S[start+index+1:stop]:
            if i==j:
                return False
    return True 
```

F(n) = F(n-1) + F(n-2)


Number of calls for each n:

n=2 ---> 1 (check n<=1) + c0 + c1 = 3 

n=3 ---> 1 + c2 + c1 = 1 + 3 + 1 = 5 

n=4 ---> 1 + c3 + c2 = 1 + 5 + 3 = 9 

n=5 ---> 1 + c4 + c3 = 1 + 9 + 5 = 15 

Therfore, c4 > 2*c2 and c5 > 2*c3. Thus, cn > 2^{n/2}

```python
def good_fibonacci(n):
    '''Return pair of Fibonacci numbers, F(n) and F(n-1)'''
    if n<=1:
        return (n,0)
    else:
        (a,b) = good_fibonacci(n-1)
        return (a+b,a)
```

```python
def good_fibonacci_while(n):
    if n<=1:
        return (n,0)
    else:
        a,b = 1,0
        for i in range(2,n+1):
            a,b = a+b,a
        return (a,b)
```
