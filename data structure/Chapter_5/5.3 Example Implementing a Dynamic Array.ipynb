## Dynamic Array 

The class allows us to add elements to the list, with no apparent limit on the overall capacity of the list. 

*Algorithm*:

1. Allocate a new array B with larger capacity (len(B)>n)

2. Set B(i) = A(i) for i = 0,1,...,n-1 for the current n numbers 

3. Set A = B 

4. Insert new element in the new array 

```python
import ctypes ## provides low-level array

class DynamicArray:
    def __init__(self):
        self._n = 0 
        self._capacity = 1
        self._A = self._make_array(self._capacity)
    
    def __len__(self):
        """Return number of elements stored in the array"""
        return self._n 
    
    def __getitem__(self,k):
        if not 0<= k < self._n:
            raise IndexError("invalid index")
        return self._A[k]
    
    def append(self,obj):
        """Add object to the end of the array"""
        if self._n = self._capacity: # no more room 
            self._resize(2*self._capacity) # dobule capacity 
        self.A[self._n] = obj 
        self._n += 1
    
    def _resize(self,c):
        B = self._make_array(c)
        for k in range(self._n):
            B[k] = self._A[k]
        self._A = B 
        self._capacity = c 
    
    def _make_array(self,c):
        return (c*ctypes.py_object)()
```

## Simple Cryptography 

Caesar cipher : replace each letter in a message with the letter that is a certain number of letters after it in the alphabet 

```python
original_index = 23 
print ("original chr : ",chr(original_index + ord("A")))
shift = 2 
print ("left shift num: ", shift)
print ("shifted chr :",chr((original_index+shift)%26 + ord("A")))

print ()
original_index = 2
print ("original chr : ",chr(original_index + ord("A")))
shift = -4
print ("left shift num: ", shift)
print ("shifted chr :",chr((original_index+shift)%26 + ord("A")))
```

```python
class CaesarCipher:
    """Class for doing encryption and decryption using a Caesar cipher"""

    def __init__(self,shift):
        encoder = [None] * 26 
        decoder = [None] * 26 
        for k in range(26):
            encoder[k] = chr( (k+shift)%26 + ord("A"))
            decoder[k] = chr( (k-shift)%26 + ord("A"))
        self._forward = "".join(encoder)
        self._backward = "".join(decoder)
    
    def encrypt(self,message):
        return self._transform(message,self._forward)
    
    def decrypt(self,message):
        return self._transform(message,self._backward)
    
    def _transform(self,message,reference):
        return "".join(reference[ord(ele)-ord("A")] for ele in list(message) if ele.isupper())
```

```python
cipher = CaesarCipher(3)
message = "THE EAGLE IS IN PLAY; MEET AT JOE'S"
coded = cipher.encrypt(message)
print ("Secret: ", coded)
answer = cipher.decrypt(coded)
print ("Message: ", answer)
```
